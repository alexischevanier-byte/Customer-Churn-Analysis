# 📊 Customer Churn Analysis

## 🎯 Objective
Analyze customer data to identify patterns and factors that influence customer churn.

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("WA_Fn-UseC_-Telco-Customer-Churn.csv")
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')
df.dropna(inplace=True)
```

```python
df['Churn'].value_counts().plot(kind='bar')
plt.title("Churn Distribution")
plt.show()
```

```python
pd.crosstab(df['Contract'], df['Churn']).plot(kind='bar')
plt.title("Churn by Contract Type")
plt.show()
```

```python
plt.figure()
plt.scatter(df['MonthlyCharges'], df['tenure'])
plt.title("Monthly Charges vs Tenure")
plt.xlabel("Monthly Charges")
plt.ylabel("Tenure")
plt.show()
```

```python
df[df['Churn'] == 'Yes']['tenure'].hist()
plt.title("Tenure of Churned Customers")
plt.show()
```

## 🔍 Insights

- Customers with month-to-month contracts have higher churn rates
- Higher monthly charges are associated with increased churn
- Customers with low tenure are more likely to churn

## 🚀 Conclusion

Customer churn is strongly influenced by contract type, tenure, and monthly charges.

Companies should focus on retaining customers with short-term contracts
and high monthly fees.
